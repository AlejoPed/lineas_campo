```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# Datos de entrada
x = np.array([2, 4.5, 6.8, 9.7, 11.5, 14, 16.8, 19, 21.2, 23.5])
y = np.array([5.9, 5, 4.8, 4.5, 4.6, 5.3, 5.9, 7.7, 9.7, 11.5])
voltajes = np.array([1.73, 3.68, 4.9, 6, 7.1, 8.66, 10.4, 13, 17.15, 19.85])

# Puntos medidos
plt.scatter(x, y, c=voltajes, cmap='viridis', label='Puntos Medidos')
plt.colorbar(label='Voltaje')

# Unir los puntos con curvas suaves
spl = make_interp_spline(np.arange(len(x)), np.column_stack([x, y]), k=3)
xy_smooth = spl(np.linspace(0, len(x)-1, 1000))
x_smooth, y_smooth = xy_smooth[:, 0], xy_smooth[:, 1]
plt.plot(x_smooth, y_smooth, 'r-', label='Curva Suave')

# Etiquetas y leyenda
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Distribuci√≥n de Voltajes en el Espacio')
plt.legend()
plt.show()
```
